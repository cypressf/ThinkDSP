```python
from __future__ import division
import thinkdsp
import thinkplot
import numpy
wave = thinkdsp.read_wave('100475__iluppai__saxophone-weep.wav')
wave.unbias()
wave.normalize()
segment = wave.segment(start=0.76, duration=0.05)
segment.plot()
segment.make_audio()
```

```python
def autocorrelate(wave):
    correlations = numpy.correlate(wave.ys, wave.ys, mode='same')
    n = len(correlations)
    lengths = range(n, n//2, -1)
    return correlations[n//2:] / lengths
```

```python
half = autocorrelate(segment)
thinkplot.plot(half)
```

```python
def find_fundamental(wave):
    sums = []
    for i, magnitude in enumerate(wave):
        if i == 0: continue
        total = 0
        for j in range(i, len(wave), i):
            total += wave[j]
        sums.append(total)
    thinkplot.plot(sums)
    thinkplot.config(yscale='log')
    period = sums.index(max(sums)) / segment.framerate
    frequency = 1 / period
    return frequency
```

```python
fundamental_freq = find_fundamental(half)
fundamental_freq
```

```python
spectrum = segment.make_spectrum()
spectrum.band_stop(low_cutoff=fundamental_freq - 6, high_cutoff=fundamental_freq + 6)
spectrum.plot()
```

```python
segment2 = spectrum.make_wave()
segment2.plot()
```

```python
new_autocorrelation = autocorrelate(segment2)
thinkplot.plot(new_autocorrelation)
```

```python
find_fundamental(new_autocorrelation)
```
